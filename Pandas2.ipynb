# 다변량 데이터로 다변량 그래프 그리기-산점도 그래프

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
#recode_sex라는 함수를 만들고 인자를 sex로 줫다.
#sex가 female이면 0을 반환하고
#sex가 그 이외의것 male이면 1을 반환하도록 한다.
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
#tips의 sex컬러 열이 tips의 sex열에 recode_Sex함수가 적용 되었을때
#1이면 그에 맞는 sexcolor를 1 or 0으로 해라
```

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 10,    #s 는 점의 크기 size
    c=tips['sex_color'],    #c 는 점의 색상을 가리키빈다.
    alpha=0.5)              #alpha 점의 투명도

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

# 밀집도 그래프

```python
import seaborn as sns

tips=sns.load_dataset('tips')

ax = plt.subplots()
ax = sns.distplot(tips['total_bill']) 
#밀집도가 그려진 그래프 만약에 밀집도 그래프를 제외하고 싶다면
#kde라는 인잣값을 flase로 변환합니다
#밑의 예시랑 같습니다.
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax=plt.subplots() 
ax=sns.distplot(tips['total_bill'], hist=False) 
#만약에 뒤에거를 지우고 밀집도그래프만 나타내고 싶으면
#hist를 false를 주면 없어진다.

ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
```

```python
ax=plt.subplots() 
ax=sns.distplot(tips['total_bill'], rug=True) 
#만약에 양탄자 그래프를 같이 그리고 싶다면 rug=true

ax.set_title('Total Bill Histogram with Density and Rug Plot') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')

#여기까지 밀집도 그래프중에
#ax=plt.subplots() 으로 서브플롯을 그려주고
#ax=sns.distplot(tips['열이름'],ked,rug,hist)
```

```python
ax=plt.subplots() 
ax=sns.countplot('day', data=tips) 
#이산 값을 나타내는 countplot을 그린다.
#ax=sns.countplot(데이터 열의이름 , data=데이터프레임 명시)
ax.set_title('Count of days') 
ax.set_xlabel('Day of the Week') 
ax.set_ylabel('Frequency') 
print(tips)
```

# 산점도 그래프와 히스토그램을 한번에 그려주는 jointplot 메서드

# 관계그래프 그리기

```python
fig=sns.pairplot(tips)
#대칭 그래프를 그리며 단점으로는 중복된 정보가 표현 됩니다.

#가장 간단한 호출은 scatterplot()각 변수 쌍 histplot()과 대각선을
#따라 주변 플롯에 사용합니다.
#scatter와 hist를 사용
```

```python
pair_grid=sns.PairGrid(tips) 
#생성자를 호출하면 데이터 집합의 숫자 변수에 
#해당하는 각 행과 하나의 열이있는 빈 서브 플롯 그리드가 설정됩니다.
#penguins = sns.load_dataset("penguins")
#g = sns.PairGrid(penguins)

pair_grid=pair_grid.map_upper(sns.regplot) 
#우상단 그래프
pair_grid=pair_grid.map_lower(sns.kdeplot)
#좌하단 그래프
pair_grid=pair_grid.map_diag(sns.distplot,rug=True)  #대각선은 히스토 그램으로 바꾼다.
#대각선은 히스토그램으로 양탄자 그래프가 그려지도록.
plt.show()
#중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체한다.
```

# seaborn라이브러리로 그래프스타일 설정하기

```python
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```

```python
sns.set_style('whitegrid') 
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```

```python
fig = plt.figure() 
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks'] 
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style) 
        
fig.tight_layout() 
```
